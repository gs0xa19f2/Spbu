# Импортируем библиотеки

```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

# Импортируем датасет - Global Terror

```python
data = pd.read_csv("GlobalTerror.csv")
data.fillna(0).head(4)
```

# Выбираем данные
*Некоторые данные слишком малы и не значимы*
<font size = "4"><font color = "grey"><p>
Extended - if terract had been continuing more than 24 hours
</p></font></font>

```python
parameters = [  "iyear", "imonth", "iday", "extended",
                "country", "region","suicide",
                "attacktype1", "targtype1","success",
                "targsubtype1", "natlty1", "nperps",
                "weaptype1", "weapsubtype1"
                ]
new_data = data[parameters].fillna(0)
new_data.head(-5)
```

# Визуализация

<font size = "6"><font color = "Grey">Визуализация (extended, targtype1, country)</font></font>

```python
df = new_data.sample(frac = 0.2)
sb.scatterplot(
    hue = df["weaptype1"],
    x = df["targtype1"],
    y = df["country"]
)
```

*Замечаем, что определенные типы объектов атакованы с применением определенных видов оружия, например в типе объекта 20, 21, 14*

<font size = "6"><font color = "Grey">Визуализация (iyear, nperps)</font></font>

```python
df = new_data.sample(frac = 1)
fig, ax = plt.subplots()
sb.scatterplot(
    data=df, 
    x="targtype1", 
    y="region", 
    hue="success", 
    size="nperps",
    sizes = (20, 2000)
);
ax.legend().set_visible(False)
```

*Регион, количество террористов и тип атакуемого объекта сильно коллерируют*

# Тренировка модели (DecisionTreeClassifier)

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
import time
X = new_data.drop("targtype1", axis = 1)
y = new_data["targtype1"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2)
tree = DecisionTreeClassifier()
start = time.time()
tree = tree.fit(X_train, y_train)
print("Train: ",time.time() - start)
start = time.time()
y_kmeans = tree.predict(X_test)
print("Predict:", time.time() - start)
tree.n_classes_
print(f"Accuracy: {tree.score(X_test, y_test)}")
```

*Количество атакуемых объектов = 22, точность предсказания типа атакуемого объекта = 98%*
<p></p>
Наиболее лучшим способом обучения подошел именно способ DecisionTreeClassifier, дающий 98% точность, в отличии от KMeans, который не давал полезных результатов.
